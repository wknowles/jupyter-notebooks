We can see that coordinates are as POINTS. I want them split into lat and lon columns. I would like to do that with some regex. To find lon the expression is: `(-\d+.\d+)+`.  For lat: 

```python
census = gpd.read_file('data/census_tracts.json')
```

```python
census.crs = {'init' :'epsg:4326'}
```

```python
# set crs as mercator
census = census.to_crs(epsg=3395) 
# plot choropleth on census name with red color scheme and no axis
census.plot(column='NAME10', cmap='OrRd').axis('off')
```
