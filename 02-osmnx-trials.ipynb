```python
import osmnx as ox, geopandas as gpd
%matplotlib inline
ox.config(log_file=True, log_console=True, use_cache=True)
```

```python
cdplace = ox.gdf_from_place('Brittany Farms-Highlands, PA')
ox.project_gdf(cdplace)
```

```python
ox.save_gdf_shapefile(cdplace)
city = ox.project_gdf(cdplace)
fig, ax = ox.plot_shape(cdplace, figsize=(4,4))
```

```python
# get the boundary polygons for multiple cities, save as shapefile, project to UTM, and plot
place_names = ['New Britain Township, PA',
               'Chalfont, PA',
               'Doylestown, PA']
bucksplaces = ox.gdf_from_places(place_names)
ox.save_gdf_shapefile(bucksplaces)
east_bay = ox.project_gdf(bucksplaces)
fig, ax = ox.plot_shape(bucksplaces)
```

```python
# create the street network within the highlands
G1 = ox.graph_from_place('Brittany Farms-Highlands, PA')
G1_projected = ox.project_graph(G1)
fig, ax = ox.plot_graph(G1_projected)
```
